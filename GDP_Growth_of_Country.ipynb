```python
import pandas as pd

df = pd.read_csv('gdp.csv')
df.head()
```

# **Analysing India Country**

```python
df_In = df[df['Country Name'] == 'India']
```

```python
df_In.plot(kind = 'line', x = 'Year', y = 'Value',
           figsize = (15,6),
           legend = False,
           grid = True,
           ylabel = 'GDP',
           xlabel = 'YEARS')
```

```python
data = df_In.values

gdp_change = [0]

for i in range(1,len(data)):

    prev = data[i-1][3]
    cur = data[i][3]

    gdp_change.append(round(((cur - prev) / prev)*100,2))

```

```python
df_In.assign(GDP = gdp_change)
```
