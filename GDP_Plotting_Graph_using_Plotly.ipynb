```python
import pandas as pd

import plotly.express as px
import plotly.offline as pyo
```

```python
df = pd.read_csv("gdp.csv")
df.head()
```

# **Add Column That Show Change in GDP**

```python
final_data = []

for country_name in df['Country Name'].unique():

    df_pr = df[df['Country Name'] == country_name]

    data = df_pr.values
    gdp_change = [0]

    for i in range(1,len(data)):

        prev = data[i-1][3]
        cur = data[i][3]

        gdp_change.append(round(((cur - prev) / prev)*100,2))

    df_pr = df_pr.assign(GDP = gdp_change)
    final_data.append(df_pr)

df = pd.concat(final_data, axis = 0)
```

```python
df.head()
```

```python
df_In = df[df['Country Name'] == 'India']

fig = px.line(df_In , x = 'Year' , y = 'Value', title = 'India GDP Analysis')
fig.show()
```

```python
pyo.plot(fig ,filename = 'Indian GDP.html')
```

```python
df_pr = df[df['Country Name'] == 'World']

fig = px.line(df_pr , x = 'Year' , y = 'Value', title = 'World GDP Analysis')
fig
```
